# Pandas Querying

In this worksheet we will look at how to select rows of data from a Pandas DataFrame which match a specific condition.  You may want to do this to restrict the data set for some purpose.

```python
# Import the Pandas module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv("doc.csv")
df
```

## Selecting rows with a query

Sometimes you need to select back just the rows of data that meet some condition.

For example, what if we want to select just the rows where the employment is greater than 70?


```python
df
```

```python
df[df.sub_result == 'clear']
```

## Select rows with multiple query conditions: and

Multiple conditions can be combined with "&" to mean "and".  Here both conditions must be met:

```python
df[df.sub_result == 'rejected']
```

```python
df[df.sub_result == 'suspected']
```

```python
df[(df.visual_authenticity_result == 'consider')]
```

## Select rows that contain a null
Here we select any rows that contain nulls.  This is useful when we are inspecting our data after we have loaded it in.

```python
df[(df.image_integrity_result == 'consider')]
```
